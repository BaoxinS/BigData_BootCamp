```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
credit = pd.read_csv('default of credit card clients.csv', header = 1)
```

```python
credit = credit.rename(columns={'default payment next month':'Default'})
```

```python
# In terms of credit limit, most borrowers have a credit limit at around NT$50,000.
```

```python
pd.crosstab(credit.EDUCATION, credit.Default, margins=True).style.background_gradient(cmap='autumn_r')
```

```python
print("% of default in") 
print("EDUCATION = Graduate School : ", credit.Default[credit.EDUCATION == 1].sum()/credit[credit.EDUCATION == 1].Default.count())
print("EDUCATION = University : ", credit.Default[credit.EDUCATION == 2].sum()/credit[credit.EDUCATION == 2].Default.count())
print("EDUCATION = High School : ", credit.Default[credit.EDUCATION == 3].sum()/credit[credit.EDUCATION == 3].Default.count())
print("EDUCATION = Other 0 : ", credit.Default[credit.EDUCATION == 0].sum()/credit[credit.EDUCATION == 0].Default.count())
print("EDUCATION = Other 4 : ", credit.Default[credit.EDUCATION == 4].sum()/credit[credit.EDUCATION == 4].Default.count())
print("EDUCATION = Other 5: ", credit.Default[credit.EDUCATION == 5].sum()/credit[credit.EDUCATION == 5].Default.count())
print("EDUCATION = Other 6: ", credit.Default[credit.EDUCATION == 6].sum()/credit[credit.EDUCATION == 6].Default.count())
```

```python
# We do not know what "Others" represents in Education. However, form the given info, the higher the education level, the lower the default rate.
```

```python
pd.crosstab([credit.SEX, credit.Default], credit.EDUCATION, margins=True).style.background_gradient(cmap='autumn_r')
```

```python
sns.catplot('EDUCATION','Default',hue='SEX', kind='point', data=credit)
```

```python
# Except that both men and women with Education-Others (0, 5, 6) have the approximately same default rate, for other education level, women have lower default rate than men.
```

```python
sns.catplot(x='Default', col='MARRIAGE', kind='count', data=credit);
```

```python
print("% of default in") 
print("MARRIAGE = Married : ", credit.Default[credit.MARRIAGE == 1].sum()/credit[credit.MARRIAGE == 1].Default.count())
print("MARRIAGE = Single : ", credit.Default[credit.MARRIAGE == 2].sum()/credit[credit.MARRIAGE == 2].Default.count())
print("MARRIAGE = Divorced : ", credit.Default[credit.MARRIAGE == 3].sum()/credit[credit.MARRIAGE == 3].Default.count())
print("MARRIAGE = Other : ", credit.Default[credit.MARRIAGE == 0].sum()/credit[credit.MARRIAGE == 0].Default.count())
```

```python
# Ohter marriage status has the lowest default rate. Divorced has the highest rate. Married and single are at middle, where married borrowers has higher default rate than single ones.
```

```python
sns.catplot('MARRIAGE','Default', hue= 'SEX', kind='point', data=credit)
plt.show()
```

```python
# No matter the marriage status, women always has lower default rate than men. Divorced men are more likely to default than any other borrowers.
```

```python
age_dist = sns.distplot(credit.AGE)
age_dist.set_title("Distribution of Borrowers' Ages")
```

```python
#Age: continuous numerical to 6 bins
```

```python
sns.lmplot('AGE', 'Default', data=credit)
```

```python
# The older the borrower is, the more likely he/she will default.
```

```python
sns.lmplot('AGE', 'Default', data=credit, hue='SEX')
```

```python
# Both older men and women are more likely to default.
```

```python
sns.lmplot('AGE', 'Default', data=credit, hue='MARRIAGE')
```

```python
# Older borrowers who are Single or with other marriage status are less likely to default; however, those elders who married or divorced are more likely to default.
```

```python
bal_dist = sns.distplot(credit.LIMIT_BAL)
bal_dist.set_title("Distribution of Borrowers' Credit Limits")
```

```python
sns.lmplot('LIMIT_BAL', 'Default', data=credit)
```

```python
# The higher credit limit a borrower has, the less possibility that he/she will default.
```

```python
sns.lmplot('LIMIT_BAL', 'Default', data=credit, hue='SEX')
```

```python
# Even with the same credit limits, men borrowers are more likely to default.
```

```python
sns.lmplot('LIMIT_BAL', 'Default', data=credit, hue='MARRIAGE')
```

```python
# No matter what marriage status the borrowers have, those who with higher credit limits are less likely to default.
```

```python
sns.lmplot('PAY_0', 'Default', data=credit)
```

```python
sns.lmplot('PAY_6', 'Default', data=credit)
```

```python
# The history of past payment has strong impact on predicting the default possibility. Those who delayed longer is more likely to default.
```

```python
pay_dist = sns.distplot(credit.PAY_0)
pay_dist.set_title("Distribution of Pay_0")
```

```python
# Most borrowers used revolving credits or paid in full.
```

```python
bill_dist = sns.distplot(credit.BILL_AMT6)
bill_dist.set_title("Distribution of BILL_AMT6")
```

```python
# Most borrowers has very small billing amount.
```

```python
sns.lmplot('BILL_AMT6', 'Default', data=credit)
```

```python
# The higher the bill amount is, the less possibility of defaulting. This could be relate to the fact that borrowers with higher credit limit are less likely to default.
```

```python
# Pay amount is in line with billing amount.
```

```python
sns.lmplot('PAY_AMT6', 'Default', data=credit)
```

```python
# The less borrowers pay back to CreditOne, the more likely he/she will default.
```

```python
# From the above analysis, we find that:
    # Women borrowers are more than men borrowers, but they are less likely to default than men
    # Borrowers with higher education level are less likely to default
    # Single borrowers and those who with other marriage status are less likley to default than those who are divorced or married
    # Borrowers with higher credit limits are those who less likely to default
    # Young borrowers have lower default rate than older borrowers
    # History of past payments is a useful indicator to predict defaults: the longer a borrower delays the payment, the more likely he/she will default
```
